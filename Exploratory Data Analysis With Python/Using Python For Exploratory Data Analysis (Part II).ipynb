The following code provides statistics for number of cylinders.

```python
# naming the only column as amount
# in the original notebook, it is count
t = pd.crosstab(index=df["Cylinders"], columns="amount")

# this actually creates a new column in the dataframe
t["percentage"] = (t/t.sum())*100
print(t)
```

Cross tab allows us to count the frequency of objects that fall under 2 or more attributes.

### Question 7: How about a horizontal bar plot? Can you learn how to control the width of bars and make a plot that has thinner bars?

We can control the "width" of the bar by setting the height attribute.

Online documentation: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.barh.html

This can be a bit tricky when I first approached this problem because in the online documentation, there is an attribute called width, which is passed as the second argument but it does not control the geometric width that is asked of us in this question.

The width that is mentioned in the online documentation is how stretched out do you want the horizontal bar to go. 

```python
# the bars are much thinner with the height property 
# set as a small value 
plt.barh(t.index,t["amount"], height=.2)
```

Pie chart

### Question 8: Do you prefer bar or pie displayed chart and why?

**Answer 8:** In a pie chart, you are able to see the distribution of one component vis a vis the whole much better. You are able to see how much does one component make up the whole. It is good to compare one component to the whole. 

In a bar chart, you are able to see the shape of a distribution. You can see whether it is skewed toward one value of another. It is good to observe skewness.

In the context of MPG in this problem, I find the vertical bar chart more insightful because MPG has a lot of varied values. 

If I put it in a pie chart, I would not be able to capture as much detail as I would like because for MPG values that are not as frequent like others, it will takes up a small part of the pie and therefore, it does not capture as much attention.

Another reason is I want to see what kind of values for MPG in which most car are concentrated around rather than to compare whether one range of MPG values occupies more share than another range of MPG values.
